```python
''' hashing elements in list to and check frequency of hashed numbers 
from another list '''

#hashing using list

n=[1,2,3,3,3,2,4,5,5,6,7,7,7,7,7,7,7,8,9,10]
m=[12,3,4,5,6,7,8,9,10,11]

hash_list=[0]*11

for num in n:
    hash_list[num]+=1

for num in m:
    if num<0 or num>=len(hash_list): print(f"Number {num} : 0")
    else: print(f"Number{num} : {hash_list[num]}")


```

```python
#hasing using dictionary

n=[1,2,3,3,3,2,4,5,5,6,7,7,7,7,7,7,7,8,9,10]
m=[12,3,4,5,6,7,8,9,10,11]

hash_dict= {}


for num in n:
    if num in hash_dict:
        hash_dict[num] += 1
    else: hash_dict[num] = 1

for num in m:
    if num<0 or num>len(hash_dict): print(f"Number {num} : 0")
    else: print(f"Number {num} : {hash_dict[num]}")
    
```

```python
#character hashing

s = "abacabadabacaba"
query = "abcdefghijklmnopqrstuvwxyz"

hash_str = {}

for ch in s:
    if ch in hash_str:
        hash_str[ch] += 1
    else:
        hash_str[ch] = 1

for ch in query:
    print(f"character {ch} : {hash_str.get(ch, 0)}")
```
