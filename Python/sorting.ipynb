```python
#selection sort

arr=[3,2,6,4,7,1,8]

def selection_sort(arr):

    n=len(arr)

    for i in range (0,n):
        min_index=i

        for j in range(i+1,n):

            if arr[j]<arr[min_index]:
                min_index=j
                
        arr[i],arr[min_index]=arr[min_index],arr[i]
            
                

selection_sort(arr)
print(arr)

```

```python
# bubble sort

arr=[3,2,6,4,7,1,8]

def bubble_sort(arr):

    n=len(arr)

    for i in range (n-2, -1, -1): # i will traverse in backwards as the elements are being sorted at the end indexes
        for j in range(0, i+1):  #j is going till i-th element cause last indexed element is already sorted
            if arr[j] > arr[j+1]:
               arr[j], arr[j+1] = arr[j+1], arr[j]



bubble_sort(arr)

print(arr)


```

```python
# bubble sort - for  the case when the array is already sorted, so loop breaks after 1 iteration

arr=[1,2,3,4,5,6]

def bubble_sort(arr):

    n=len(arr)

    for i in range (n-2, -1, -1): # i will traverse in backwards as the elements are being sorted at the end indexes

        is_swap=False

        for j in range(0, i+1):  #j is going till i-th element cause last indexed element is already sorted
            if arr[j] > arr[j+1]:
               arr[j], arr[j+1] = arr[j+1], arr[j]
               is_swap=True
        if is_swap == False: return "array is already sorted"



result = bubble_sort(arr)

if result: print(result)
else: print(arr)

```

```python
# insertion sort

arr=[3,2,6,4,7,1,8]

def insertion_sort(arr):

    n=len(arr)

    for i in range(1,n):

        is_swap=False

        key = arr[i]
        j = i-1

        while j>=0 and arr[j]>key:

            arr[j+1] = arr[j]
            j-=1

            if is_swap == True: return 'array already sorted'
    

        arr[j+1] = key


result=insertion_sort(arr)

if result: print(result)
else: print(arr)




```

```python
# merge sort (recursive)

arr=[3,1,2,4,1,5,2,6,4]


def merge_array(left,right):

    result=[]
    i,j=0,0
    n,m=len(left),len(right)

    while i<n and j<m:

        if left[i] <= right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    
    #exhaustion part

    if i<n:
        result.append(left[i])
        i+=1

    if j<m:
        result.append(right[j])
        j+=1

    return result




def merge_sort(arr):

    if len(arr) <= 1: return arr

    mid=len(arr)//2

    left_arr = arr[:mid]
    right_arr = arr[mid:]

    left =  merge_sort(left_arr)
    right = merge_sort(right_arr)

    return merge_array(left,right)


res=merge_sort(arr)
print(res)



```
