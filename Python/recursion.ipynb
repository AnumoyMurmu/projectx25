```python
count=0
def greet():
    global count
    if count==4: return
    print("anything")
    count+=1
    greet()

greet()
```

```python
#tail recursionc
ount=0
def greet():
    global count
    if count==4: return
    count+=1
    greet()
    print("anything")

greet()
```

```python
#recursion using parameters
count=0
def func(x,y):
    global count
    if count==4: return
    print(x+y)
    count+=1
    func(x,y)

func(23,31)
```

```python
#recursion using parameters

def func(x,y):
    if y==0: return
    print(x+y)
    func(x,y-1)

func(4,18)

```

```python
#parameterised and functional recursion
#suppose n as input

def func(sum,i,n):
    if i>n: 
      print(sum) 
      return
    func(sum+i, i+1,n)

func(0,1,5)
```

```python
#sum of n numbers using functional recursion

def func(n):
    if n==1: return 1
    return n+func(n-1)

func(100)
```

```python
#factorial of a number using recursion

def factorial(n):
    if n==0 or n==1: return 1
    return n*factorial(n-1)

factorial(4)

#tc=o(N)
#stack space = o(N)
```

```python
#reverse an array using recursion

n=[1,2,3,4]

def reverse_array(n,l,r):
    if l>=r: return 1
    n[l], n[r] = n[r], n[l]
    reverse_array(n,l+1,r-1)

reverse_array(n,0,len(n)-1)
print(n)


```

```python
str="BIHARI"
def palindrome(str,l,r):
    if l>=r:
        print("Palindrome")
        return


    if str[l]==str[r]: palindrome(str,l+1,r-1)
    else: print("Not a palindrome")

palindrome(str,0,len(str)-1)



```

```python
#fibonachhi series

def fibonacci_sequence(n):
    if n==0 or n==1: return n
    return fibonacci_sequence(n-1)+fibonacci_sequence(n-2)

fibonacci_sequence(8)

```
