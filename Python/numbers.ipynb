```python
#reverse a number

n=3487

res=0
while n>0:
    num=n%10
    res=res*10+num
    n=n//10

res
```

```python
#count digits in a number

n=311157

count=0

while n>0:
    n=n//10
    count+=1

print(count)
```

```python
#palindrome number

n=14441
num=n
s=0
while n>0:
    a=n%10
    s=s*10+a
    n=n//10

if s==num:
    print(True)
else:
    print(False)
```

```python
n=153

num=n
s=0
while n>0:
    r=n%10
    s=s+r ** len(str(num))
    n=n//10

if s==num:
    print(True)
else:
    print(False)
```

```python
#factors of a number

n=19837
result=[]

for i in range(1,n+1):
    if n%i==0:
        result.append(i)

result
```
