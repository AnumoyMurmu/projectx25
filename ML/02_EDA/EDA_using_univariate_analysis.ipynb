```python
df=pd.read_csv('train.csv')

df.sample(5)
```

```python
import seaborn as sns
```

```python
# 1.categorical data
# a. countplot

# show numeric counts
counts = df['Survived'].value_counts().sort_index()
print('Not survived (0):', counts.get(0, 0))
print('Survived (1):', counts.get(1, 0))

# plot with counts annotated
ax = sns.countplot(x='Survived', data=df)
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')
```

```python
# counts per passenger class
pclass_counts = df['Pclass'].value_counts().sort_index()
print('Class 1:', pclass_counts.get(1, 0))
print('Class 2:', pclass_counts.get(2, 0))
print('Class 3:', pclass_counts.get(3, 0))

# plot with annotations
ax = sns.countplot(x='Pclass', data=df)
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')
```

```python
import matplotlib.pyplot as plt

# summary stats
fare_stats = df.groupby('Pclass')['Fare'].agg(['mean', 'median', 'std', 'count']).sort_index()
print(fare_stats)

# plot fare distribution by Pclass
fig, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=df, ax=ax)
# add swarm for individual points (smaller sample overlap)
sns.swarmplot(x='Pclass', y='Fare', data=df, color='0.25', size=3, alpha=0.6, ax=ax)

# annotate means
means = fare_stats['mean']
for i, pclass in enumerate(means.index):
    m = means.loc[pclass]
    ax.text(i, m + (fare_stats['std'].mean() * 0.05), f"mean={m:.2f}", ha='center', color='red', weight='bold')

ax.set_title('Fare distribution by Passenger Class (Pclass)')
ax.set_ylabel('Fare')
ax.set_xlabel('Pclass')
plt.tight_layout()
plt.show()

# numeric differences between classes (mean fares)
print("\nMean fare differences:")
print(f"Class 1 - Class 2: {means.loc[1] - means.loc[2]:.2f}")
print(f"Class 1 - Class 3: {means.loc[1] - means.loc[3]:.2f}")
print(f"Class 2 - Class 3: {means.loc[2] - means.loc[3]:.2f}")
```

```python
# plot total count by sex
fig, ax = plt.subplots(figsize=(6, 4))
counts.plot(kind='bar', ax=ax, color=['#1f77b4', '#ff7f0e'])
ax.set_title('Passenger count by Sex')
ax.set_xlabel('Sex')
ax.set_ylabel('Count')
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

```python
counts = df['Sex'].value_counts()
print('Male:', counts.get('male', 0))
print('Female:', counts.get('female', 0))

# plot with counts annotated (ensure consistent order)
ax = sns.countplot(x='Sex', data=df, order=['male', 'female'])
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')
```

```python
counts = df['Embarked'].value_counts()
print('Q', counts.get('Q', 0))
print('C', counts.get('C', 0))
print('S', counts.get('S', 0))


# plot with counts annotated (ensure consistent order)
ax = sns.countplot(x='Embarked', data=df, order=['Q', 'C', 'S'])
for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')
```

```python
# b. PieChart
# pie chart of Embarked distribution

labels = ['S', 'C', 'Q']
vals = counts.reindex(labels).fillna(0).astype(int)

fig, ax = plt.subplots(figsize=(6, 6))
explode = (0.03, 0.03, 0.03)

def make_autopct(values):
    def my_autopct(pct):
        total = values.sum()
        val = int(round(pct * total / 100.0))
        return f"{pct:.1f}%\n({val})"
    return my_autopct

colors = ['#66c2a5', '#fc8d62', '#8da0cb']
wedges, texts, autotexts = ax.pie(
    vals,
    labels=labels,
    autopct=make_autopct(vals),
    startangle=140,
    explode=explode,
    colors=colors,
    textprops={'fontsize': 10}
)

ax.set_title('Embarked distribution (count and %)')
ax.axis('equal')  # keep as circle
plt.tight_layout()
plt.show()
```

```python
# 2. numerical data
# a. histogram - always plot histogram for a numerical data

import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=50)

```

```python
# b. distplot

sns.displot(df['Age'])
```

```python
ages = df['Age'].dropna()

mean_age = ages.mean()
median_age = ages.median()

# histogram with KDE
sns.histplot(ages, bins=30, kde=True, color='#66c2a5', stat='density', alpha=0.6)

# annotate mean and median
plt.axvline(mean_age, color='red', linestyle='--', linewidth=1.5, label=f'Mean: {mean_age:.2f}')
plt.axvline(median_age, color='blue', linestyle=':', linewidth=1.5, label=f'Median: {median_age:.2f}')

plt.title('Age distribution (Histogram + KDE)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```
