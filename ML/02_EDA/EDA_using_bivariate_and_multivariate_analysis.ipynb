```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.sample(5)
```

```python
titanic=pd.read_csv('train.csv')

titanic.sample(5)
```

```python
flights=sns.load_dataset('flights')
flights.sample(5)
```

```python
iris=sns.load_dataset('iris')
iris.sample(5)
```

```python
# 1. scatterplot (numerical -numerical)
# The DataFrame is passed to 'data', and column names are passed to 'x' and 'y'
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# 1.2 scatterplot (numerical -numerical) - multivariate

sns.scatterplot(x='total_bill', y='tip', data=tips, hue=tips['sex'])
```

```python
# 1.3 scatterplot (numerical -numerical) - multivariate

sns.scatterplot(x='total_bill', y='tip', data=tips, hue=tips['sex'], style=tips['smoker'])
```

```python
# 1.4 scatterplot (numerical -numerical) - multivariate

sns.scatterplot(x='total_bill', y='tip', data=tips, hue=tips['sex'], style=tips['smoker'], size=tips['size'])
```

```python
# bar plot (numerical - categorical)

titanic.sample(5)
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic)
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic )
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic, hue=titanic['Sex'] )
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic, hue=titanic['Sex'] )
```

```python
# 3. box plot (numerical - categoriacl)

sns.boxplot(x='Sex', y='Age', data=titanic, hue=titanic['Survived'])
```

```python
# 4. distplot (numeircal - categorical)

sns.displot(x='Age', data=titanic)
```

```python
sns.displot(titanic[titanic['Survived']==0]['Age'])
sns.displot(titanic[titanic['Survived']==1]['Age'])
```

```python
# 5. heat map ( categorical - categorical)

pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean(numeric_only=True)
```

```python
titanic.groupby('Pclass').mean(numeric_only=True).plot(kind='bar')
```

```python
titanic.groupby('Sex').mean(numeric_only=True)
```

```python
titanic.groupby('Embarked').mean(numeric_only=True)
```

```python
# 6. clustermap (categorical - categorical)

# Create the contingency table
crosstab_data = pd.crosstab(titanic['Pclass'], titanic['Survived'])

crosstab_data
```
