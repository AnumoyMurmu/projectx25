```python
# convertors - change full names to short form i.e royal challenegers bengaluru to rcb

def rename(name):
    if name == 'Royal Challengers Bangalore': return 'RCB'
    else: return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1': rename})
```

```python
# na_value parameter - to change values into NaN

pd.read_csv('aug_train.csv', na_values=['Male'])

```

```python
# loading a huge dataset in chunks - divides a huge dataset into chunks

dfs=pd.read_csv('aug_train.csv', chunksize=500)

for chunk in dfs:
    print(chunk.size)
    print(chunk.shape)
    print('---------')

```
