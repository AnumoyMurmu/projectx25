#### 3.Univariate Analysis (Individual Feature Exploration)


```python
#Task 5: Numerical Distribution (sepal_length) 
#Let's see the range and common values of one numerical feature.

plt.figure(figsize=(8, 5))
sns.histplot(df['sepal_length'], kde=True, bins=15)
plt.title('Distribution of Sepal Length')
plt.show()
```

Why? A Histogram shows the distribution‚Äîhow often values fall into certain bins. The KDE line (Kernel Density Estimate) is a smooth curve that approximates the shape. This shows us the typical length of a sepal.

```python
# Task 6: Categorical Distribution (species) üè∑Ô∏è
# Our target variable is species (the type of flower).

plt.figure(figsize=(6, 4))
sns.countplot(x='species', data=df)
plt.title('Count of Each Iris Species')
plt.show()
```

Why? A Count Plot shows us the balance of our target variable. You should see that all three species are perfectly balanced (50 counts each). This is ideal for a machine learning classification problem!

#### Phase 4: Bivariate Analysis (Finding Relationships)

This is the core of EDA: discovering how one feature relates to another. Since our goal is classification (predicting the species), we compare every measurement to the species column.

```python
'''
Task 7: Numerical Feature vs. 
Categorical Target (Box Plot) üì¶

We use a Box Plot to visualize how the distribution 
of a numerical feature changes across different categorical groups.
'''

plt.figure(figsize=(8,5))
sns.boxplot(x='species', y='sepal_width', data=df)
plt.title('sepal width distribution by species')
plt.show()
```

Interpretation: Look at the median line inside each box. You should see that the Setosa species has the highest median sepal width. This tells you that wide sepals are a good way to identify Setosa flowers.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Petal Length Distribution by Species')
plt.show()
```

Interpretation: This is usually the most important plot! You'll see that the petal length for Setosa is completely separate from the other two species (Virginica and Versicolor), making it an extremely good predictor.

```python
'''
Task 8: All Numerical Features Together (Pair Plot) 

A Pair Plot is the single most powerful visualization 
tool for a small dataset like Iris. It creates a grid:

    1.Diagonal: Histograms/KDE plots (Univariate view).

    2.Off-Diagonal: Scatter plots for every pair of features (Bivariate view).
'''
#Create the Pair Plot: We use the hue argument to color the points based on the target variable (species).

sns.pairplot(df, hue='species', height=2.5)
plt.suptitle('Pair Plot of Iris Features', y=1.02)
plt.show()
```

Interpretation: Look at the scatter plots. You are looking for separation. The goal is to find a plot where the three colors (species) are cleanly separated into different groups. The plot of Petal Length vs. Petal Width usually shows the clearest separation.

Based on the Box Plots (Task 7) and the Pair Plot (Task 8), you can draw concrete conclusions about how to classify Iris flowers.
